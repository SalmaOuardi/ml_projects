```python
import pandas as pd

# Load processed data
train_df = pd.read_csv("../data/processed/clean_train.csv")
test_df = pd.read_csv("../data/processed/clean_test.csv")

# Check missing values
print("ðŸ”¹ Missing Values in Train:\n", train_df.isnull().sum().sum())
print("ðŸ”¹ Missing Values in Test:\n", test_df.isnull().sum().sum())

# Check transformed skewed features
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(train_df["SalePrice"], kde=True, bins=30)
plt.title("Transformed SalePrice Distribution")
plt.show()
```

```python
from scipy.stats import skew

# Compute skewness after transformation
final_skew = skew(train_df["SalePrice"])
print(f"ðŸ”¹ Final Skewness: {final_skew:.4f}")
```
